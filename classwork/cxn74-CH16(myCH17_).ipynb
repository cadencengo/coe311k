```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the parameters
k = 1.0        # Constant in the equation (e.g., spring constant)
h = np.array([0.01, 0.02, 0.03, 0.05, 0.1, 0.2, 0.3, 0.5]) # Step size
x_start = 0    # Start of the interval
x_end = 10     # End of the interval
y_0 = 1.0      # Initial condition for y
v_0 = 0.0      # Initial condition for dy/dx

for j in range (len(h)):
    # Create arrays to store x, y, and v values
    x_values = np.arange(x_start, x_end + h[j], h[j])
    y_values = np.zeros(len(x_values))
    v_values = np.zeros(len(x_values))
    # Set initial conditions
    y_values[0] = y_0
    v_values[0] = v_0

    # Euler's Method loop
    for i in range(1, len(x_values)):
        # Current values
        x_n = x_values[i - 1]
        y_n = y_values[i - 1]
        v_n = v_values[i - 1]    
        # Update equations
        y_values[i] = y_n + h[j] * v_n
        v_values[i] = v_n + h[j] * (-k * y_n)
        
    plt.plot(x_values, y_values, label=f"y(x), h={h[j]}")
    #plt.plot(x_values, v_values, label=f"v(x), h={h[j]}", linestyle='--')
    
# Plot the results
plt.xlabel('x')
plt.ylabel('y and v')
plt.title('Solution of d²y/dx² = -ky using Euler\'s Method')
plt.legend()
plt.grid(True)
plt.show()
```

The output becomes unstable as step size (h) increases, specifically when h = 0.3 and the line is not as smooth at the end. This instability is more obvious in the larger h = 0.5, where the solution has large, unreasonable oscillations.
