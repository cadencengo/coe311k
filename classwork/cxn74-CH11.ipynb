```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline
```

```python
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 1, 0, 1, 0, 1])

# Polynomial fit
degree = 5
coefficients = np.polyfit(x, y, degree)
polynomial = np.poly1d(coefficients)
y_pred = polynomial(x)

# Cubic spline
cs = CubicSpline(x, y, bc_type='natural')
x_interp = np.linspace(0, 5, 100)
y_interp = cs(x_interp)

plt.scatter(x, y, label='Data points')
plt.plot(x, y_pred, color='red', label=f'Polynomial fit of degree {degree}')
plt.plot(x_interp, y_interp, 'b-', label="Cubic Spline Interpolation")
plt.title('Polynomial vs. Cubic Spline Fit')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
cs_clamped = CubicSpline(x, y, bc_type=((1, 0), (1, 0)))
y_clamped_interp = cs_clamped(x_interp)

plt.scatter(x, y, label='Data points')
plt.plot(x, y_pred, color='red', label=f'Polynomial fit of degree {degree}')
plt.plot(x_interp, y_clamped_interp, 'b-', label="Clamped Spline Interpolation")
plt.title('Polynomial vs. Clamped Spline Fit')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
