```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the function and the domain
f = lambda x: np.sin(x)
a, b = 0, 2 * np.pi # Interval
N = 50 # Number of sample points

# Create a uniform grid
x = np.linspace(a, b, N)
y = f(x)

# Step size
h = (b - a) / (N - 1)

# Central Difference Approximation for the derivative
df_dx = (f(x + h) - f(x - h)) / (2 * h)

# Forward Difference Approximation
df_dx2 = (f(x + h) - f(x)) / h

# Backward Difference Approximation
df_dx3 = (f(x) - f(x -h)) / h

# Plot the approximation and the true derivative
plt.plot(x, np.cos(x), 'r', label='True Derivative')
plt.plot(x, df_dx, 'g', label='Central Difference Approximation')
plt.plot(x, df_dx2, 'b', label='Foward Difference Approximation')
plt.plot(x, df_dx3, 'purple', label='Backward Difference Approximation')
plt.legend()
plt.ylabel("f'(x)")
plt.xlabel("x")
plt.title("Comparing Methods of Approxmating Derivative of f(x)=sin(x)")
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x, np.cos(x), 'r', label='True Derivative')
plt.legend()
plt.ylabel("f'(x)")
plt.xlabel("x")

plt.subplot(2,2,2)
plt.plot(x, df_dx, 'g', label='Central Difference Approx.')
plt.plot(x, np.cos(x), 'r--', label='True Derivative')
plt.legend()
plt.ylabel("f'(x)")
plt.xlabel("x")

plt.subplot(2,2,3)
plt.plot(x, df_dx2, 'b', label='Foward Difference Approx.')
plt.plot(x, np.cos(x), 'r--', label='True Derivative')
plt.legend()
plt.ylabel("f'(x)")
plt.xlabel("x")

plt.subplot(2,2,4)
plt.plot(x, df_dx3, 'purple', label='Backward Difference Approx.')
plt.plot(x, np.cos(x), 'r--', label='True Derivative')
plt.legend()
plt.ylabel("f'(x)")
plt.xlabel("x")

plt.show()
```

The central difference approximation is the closest to the true derivative.

The forward difference approximation begins as a slight overestimate as the derivative decreases (and the function is concave down), and then is an underestimate as the derivative increases (the function is concave up).

The backward difference approximation is an underestimate as the derivative decreases, and then is an overestimate as the derivative increases.
