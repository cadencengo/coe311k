```python
import numpy as np
import random
import time
import matplotlib.pyplot as plt
```

```python
def naive_matrix_multiplication(A, B):
    n = len(A)
    C = [[0] * n for _ in range(n)]
    
    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
    return C
```

```python
time_naive = []
time_np = []

for i in range(0, 8):
    n = 3*(2**i)
    
    A = np.random.rand(n, n)
    B = np.random.rand(n, n)
    
    start_naive = time.time()
    C = naive_matrix_multiplication(A, B)
    end_naive = time.time()
    time_naive.append(end_naive-start_naive)
    
    start_np = time.time()
    np.dot(A, B)
    end_np = time.time()
    time_np.append(end_np-start_np)

print("custom matrix multiplication:", "\n", time_naive, "\n")
print("numpy matrix multiplication:\n", time_np)
```

```python
x = np.arange(1, 9)

plt.plot(x, time_naive, color='r', label='custom matrix multiplication')
plt.plot(x, time_np, color='b', label='numpy matrix multiplication')
plt.title("Matrix Multiplication: Custom vs. Numpy Function Runtimes")
plt.ylabel("time to multiply matrix pair (seconds)")
plt.xlabel("matrix pair number")
plt.legend()
plt.show()
```

For both the custom matrix multiplication function and the numpy matrix multiplication function, as the sizes of the matrix pair gets larger, the time it takes to multiply them increases. However, the runtime for the custom matrix multiplier increases much faster than for the numpy function, meaning that the built in numpy function is more efficient.
